```python
# Keep only selected columns from the food_deserts DataFrame
food_deserts_filtered = food_deserts[[
    'Zip Code',             # Will rename to 'Zip Code'
    'food_desert',
    'State',
    'County',
    'Pop2010',
    'PovertyRate',
    'MedianFamilyIncome',
    'lasnap1'
]].copy()

# Rename ZIP column for consistency
food_deserts_filtered.rename(columns={'ZIP': 'Zip Code'}, inplace=True)

# Display preview
print("🗺️ Filtered Food Deserts Data:")
display(food_deserts_filtered.head())
```

#### **Data Cleaning**

```python
# Show all column names in the dataset
print(food_pantries.columns.tolist())
```

```python
# Filter the specified columns from the food_pantries dataset
food_pantries_filtered = food_pantries[[
    'name',
    'addressLocality',
    'addressRegion',
    'Zip Code',
    'geolocation.coordinates'
]].copy()

# Preview the result
print("✅ Filtered Food Pantries DataFrame:")
display(food_pantries_filtered.head())


# Filter rows where Zip Code is 12204
sample = food_pantries[food_pantries['Zip Code'] == 12204]

# Display the result
display(sample)
```

```python
farmers_market['Program Type'] = 'farmersMarket'

farmers_market = farmers_market[[
    'Market Name',
    'Street Address',
    'Latitude',
    'Longitude',
    'Location Point',
    'Zip Code',
    'Program Type'
]].copy()

print("🧺 Farmers Market:")
display(farmers_market.head())
```

```python
# Merging Data to One

# ✅ Merging only Food Pantries and Farmers Markets

# Set Program Type
food_pantries_filtered['Program Type'] = 'foodPantry'
farmers_market['Program Type'] = 'farmersMarket'

# Rename for consistency
food_pantries_renamed = food_pantries_filtered.rename(columns={
    'name': 'Name',
    'Zip Code': 'Zip Code',
    'geolocation.coordinates': 'Location Point'
})[['Name', 'Zip Code', 'Location Point', 'Program Type']]

farmers_market_renamed = farmers_market.rename(columns={
    'Market Name': 'Name',
})[['Name', 'Zip Code', 'Location Point', 'Program Type']]

# Concatenate both intervention types
interventions = pd.concat([food_pantries_renamed, farmers_market_renamed], ignore_index=True)

# ✅ Preview merged interventions (only food pantries and farmers markets)
display(interventions.head(200))
```

```python
# Select only relevant food desert columns to merge
food_desert_columns = ['Zip Code', 'food_desert', 'PovertyRate', 'MedianFamilyIncome']
food_deserts_selected = food_deserts_filtered[food_desert_columns].copy()

# Merge with interventions based on Zip Code
merged_df = interventions.merge(food_deserts_selected, on='Zip Code', how='left')

# Preview the result
# display(merged_df.head(500))
# View rows from index 20 to 149 (Python indexing is zero-based and exclusive at the end)
display(merged_df.iloc[180:250])

```
